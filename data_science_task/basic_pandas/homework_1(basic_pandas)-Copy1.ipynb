```python
import pandas as pd
import numpy as np

```

```python
adult_data = pd.read_csv('https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/adult.data.csv')
adult_data
```

```python
# 1. Сколько мужчин и женщин (признак sex) представлено в этом наборе данных?

adult_data['sex'].value_counts()
```

```python
# 2. Каков средний возраст (признак age) женщин?

mean_Female_age =  adult_data[adult_data['sex'] == 'Female']['age'].mean()
np.round( mean_Female_age, 2)
```

```python
# 3. Какова доля граждан Германии (признак native-country)?

from_German = adult_data[adult_data["native-country"] == 'Germany']

# from_German_count = len(from_German.index)
from_German_count = from_German.shape[0]

all_count = adult_data.shape[0]

from_German_count / all_count * 100
```

```python
# 4-5. Каковы средние значения и среднеквадратичные отклонения(std) возраста тех, кто получает более 50K в год (признак salary) 
#               
more_then_50K = adult_data[adult_data['salary'] == '>50K' ]
more_then_50K.describe()
```

```python
# и тех, кто получает менее 50K в год?
less_then_50K = adult_data[adult_data['salary'] == '<=50K' ]
less_then_50K.describe()
```

```python
# 6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters или Doctorate)

condition  = pd.Series(["Bachelors", "Prof-school", "Assoc-acdm", "Assoc-voc", "Masters" , "Doctorate"])
more_then_50K['education'].isin(condition).all()
```

```python
# 7. Выведите статистику возраста для каждой расы (признак race) и каждого пола. Используйте groupby и describe. Найдите таким образом максимальный возраст мужчин расы Amer-Indian-Eskimo.
adult_data.groupby(["race" , "sex"]).describe()
```

```python
# 8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.

marital_true_share = adult_data[adult_data["marital-status"].map(lambda x : x.startswith("Married"))]["salary"].value_counts(normalize=True)[">50K"]

marital_false_share = adult_data[adult_data["marital-status"].map(lambda x : not (x.startswith("Married")))]["salary"].value_counts(normalize=True)[">50K"]

marital_true_share > marital_false_share

```

```python
#  9. Какое максимальное число часов человек работает в неделю (признак hours-per-week)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много? *

hours_per_week_series = adult_data["hours-per-week"]
max_hours_per_week = hours_per_week_series.max()
print("max_hours_per_week: " , max_hours_per_week)

work_max_hours = adult_data[hours_per_week_series == max_hours_per_week]
work_max_hours_count = len(work_max_hours)
print("work_max_hours_count: " , work_max_hours_count)

more_than_50K_percent  = work_max_hours["salary"].value_counts(normalize=True)[">50K"] * 100
print("more_than_50K_percent: " , more_than_50K_percent)
```

```python
# 10. Каково среднее время работы (hours-per-week) зарабатывающих мало и много (salary) для Японии? *
adult_data[adult_data["native-country"] == "Japan"].groupby("salary")["hours-per-week"].mean()
```
