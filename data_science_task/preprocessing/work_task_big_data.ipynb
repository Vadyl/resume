```python
import pandas   as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import re
from datetime import datetime
movies_metadata = pd.read_csv('C:\programmer_shit\data_salo_files\work\movies_metadata.csv' , low_memory=False)
movies_metadata = movies_metadata.drop_duplicates()
movies_metadata.head(5)
```

```python
ratings_small = pd.read_csv('C:/programmer_shit/data_salo_files/work/ratings_small.csv')
ratings_small
```

```python


def get_genres_str(genre):
     
    ind_info = [m.start() for m in re.finditer("'name': ", genre)]
    ind_info = [i + 9 for i in ind_info ]


    genres_str = ""
    for ind_info_values in ind_info:

        end = genre[ind_info_values:]
        end_i = end.find("'")
        class_info = genre[ind_info_values : end_i + ind_info_values]
        genres_str += (class_info + ", ")
        
        
    return genres_str


def get_YearsSpent(timestamp):

    dt_object = datetime.fromtimestamp(timestamp)
    dt_now = datetime.now()

    return dt_now.year - dt_object.year



def get_FavoriteGenre(userId):
    

    all_Genres_str = ""
    for i in ratings_small.query("rating >= 4 and userId ==" + str(userId)).movieId:

        for mp in MovieProfile:
            try:
                a = int(mp["MovieId"])
                b = int(i)
                if a == b:
                    all_Genres_str += mp["Genres"]
            except:
                continue



    all_Genres = all_Genres_str.split(", ")



    all_Genres_dict = {i:all_Genres.count(i) for i in all_Genres}


    max_v = 0
    max_k = ""
    for k,v in all_Genres_dict.items():
        if max_v < v:
            max_k = k
            max_v = v

    return  max_k      
```

```python
MovieProfile = []

for i in range(45449):
    
    try:
        MovieName = movies_metadata.original_title[i]
    except:
        continue
    
    MovieId = movies_metadata.id[i]
    Genres = get_genres_str(movies_metadata.genres[i])
    
    try:
        MovieRating = ratings_small.groupby("movieId")["rating"].mean()[int(MovieId)]
    except:
        MovieRating = 0
    
    MovieProfile.append({"MovieId" : MovieId, "MovieName" : MovieName,
                       "MovieRating" : MovieRating, "Genres" : Genres })
MovieProfile
```

```python
UserProfile = []

NumberOfMarks_list = ratings_small.groupby("userId")["rating"].count()
first_mark_list = ratings_small.groupby("userId")["timestamp"].min()
AvgMark_list = ratings_small.groupby("userId")["rating"].mean()



for i in range(100):
    UserId = ratings_small.userId.unique()[i]
    NumberOfMarks = NumberOfMarks_list[int(UserId)]
    YearsSpent = get_YearsSpent(first_mark_list[int(UserId)])
    AvgMark = round(AvgMark_list[int(UserId)] , 2)
    AvgTimeBetweenMarks = round(ratings_small.query("userId == " + str(UserId) ).timestamp.sort_values().diff().mean()) // 86400
    FavoriteGenre = get_FavoriteGenre(UserId)
    
    UserProfile.append({"UserId" : UserId, "NumberOfMarks" : NumberOfMarks,"YearsSpent" : YearsSpent,"AvgMark" : AvgMark, 
                        "AvgTimeBetweenMarks" : AvgTimeBetweenMarks,"FavoriteGenre" :  FavoriteGenre } )
                   
UserProfile
```
