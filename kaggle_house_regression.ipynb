```python
import numpy as np
import pandas as pd
```

```python
oringinal = pd.read_csv('train.csv' , index_col='Id')
oringinal_test = pd.read_csv('test.csv', index_col='Id')
```

```python
oringinal
```

```python
test_data = oringinal_test.copy()
train_data = oringinal.copy()
```

```python
oringinal.plot.scatter(x='GrLivArea' , y='SalePrice')

```

```python
oringinal["GrLivArea"].sort_values(ascending=False)
```

```python
train_data = train_data[train_data["GrLivArea"] < 4476]
```

```python
train_data
```

```python
target = train_data['SalePrice']
```

```python
train_data=train_data.drop("SalePrice" , axis = 1)
```

```python
train_test = pd.concat([train_data, test_data])
train_test
```

```python
count_of_nan = train_test.isna().sum().sort_values(ascending=False)
count_of_nan[count_of_nan > 0]
```

```python
train_test.drop(count_of_nan[count_of_nan > 0].index ,  axis = 1 , inplace=True )
```

```python
train_test
```

```python
(np.abs(train_test.corr()) > 0.85).sum()
```

```python
train_test_dummies = pd.get_dummies(train_test)
train_test_dummies
```

```python
train_test_dummies+=1
train_test_dummies_ln = np.log(train_test_dummies)
train_test_dummies_ln
```

```python
target_log = np.log(target)
target_log
```

```python
X = train_test_dummies_ln.iloc[:target.count()]
X_test = train_test_dummies_ln.iloc[target.count():]
```

```python
X_test
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, target_log, train_size=0.85, random_state=4246)
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_val_pred = linear_regressor.predict(X_val)
mean_squared_error(y_val, y_val_pred)
```

```python
test_pred = linear_regressor.predict(X_test)
test_pred
```

```python
final_preds_linear =  np.exp((test_pred)) 
final_preds_linear
```

```python
import xgboost
xgboost_regressor = xgboost.XGBRegressor()
xgboost_regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_val_pred = xgboost_regressor.predict(X_val)
mean_squared_error(y_val, y_val_pred)
```

```python
test_pred_xgboost = xgboost_regressor.predict(X_test)
test_pred_xgboost
```

```python
final_preds_xgboost =  np.exp((test_pred_xgboost)) 
final_preds_xgboost
```

```python
output = pd.DataFrame({'Id': X_test.index,
                      'SalePrice': final_preds_linear})
output.to_csv('submission_lin.csv', index=False)
```
